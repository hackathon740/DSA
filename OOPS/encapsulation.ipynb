```python
class Bank:
    def __init__(self,name:str,balance:int):
        self.name:str = name
        #private
        self.__balance:int = balance

    def get_balance(self):
        print(f"Current Balance = {self.__balance}")

    def set_balance(self,new_amount):
        self.__balance = new_amount
    
    def __isServerLive(self):
        return True

    def deposit(self,amount:int):
        if self.__isServerLive() == True:
            self.__balance +=amount
            print(f"amount deposited , current balance = {self.__balance}\n")
        else:
            print("Server is down")
        
    def withdraw(self,amount:str):
        if amount>self.__balance:
            print("not enough money in the bank\n")
        else:
            self.__balance -= amount
            print(f"amount withdraw , current balance = {self.__balance}\n")

acc = Bank(name="abhishek" , balance=10000)
acc.get_balance()
acc.set_balance(new_amount=6000)
acc.deposit(amount = 2000)
acc.withdraw(amount = 5000)
```
