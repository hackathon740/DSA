```python
class Students:

    name = " tarun "
    roll = 2
    subject = " "

    def report(self,a,b,c):
        self.name = a
        self.roll = b
        self.subject = c
        print(f"self :- {self}")
        print(f"the class boy {self.name} made a huge success with a {self.roll} number in this {self.subject} subject")

s1 = Students()

print(s1.name)
print(s1.roll)
s1.report(a="ankush",b=10 , c="maths")

print(s1)


s2 = Students()
s2.report(b=60,c="english",a="avinash")
```

```python
class Green:

    def __init__(self,name:str,age:int,gender:str):
        self.name=name
        self.age=age
        self.gender=gender

    def display(self) -> None:
        print(f"my name is {self.name} and my age is {self.age} and i am a {self.gender}")

g1=Green("abhishek",25,"male")
g1.display()
```

```python
class Movie:

    def __init__(self, movie_name:str , total_seats:int , ticket_price:int):
        self.movie_name = movie_name
        self.total_seats = total_seats
        self.booked_seats = 0
        self.ticket_price = ticket_price

    def book_tickets(self,no_of_tickets:int):
        if no_of_tickets >  self.total_seats - self.booked_seats:
            print("Sorry ! no enough seats")
        else:
            self.booked_seats += no_of_tickets
            self.total_seats -= no_of_tickets
            print(f"your ticket is booked")
            print(f"Total Price of the tickets :- {self.ticket_price*no_of_tickets}\n")

    def status(self):
        print(f"movie name :- {self.movie_name}")
        print(f"Booked Seats :- {self.booked_seats}")
        print(f"Seats Available :- {self.total_seats}")

movie = Movie(movie_name="Krish",total_seats=10,ticket_price=160)
movie.book_tickets(17)
movie.status()
```
