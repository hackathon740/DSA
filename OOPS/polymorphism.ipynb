```python
class Animal:

    def __init__(self,name:str,age:int):
        self.name=name
        print(f"Name of the animale :- {self.name}")
        self.age=age
        print(f"Name of the animale :- {self.age}")
        
    def eat(self):
        print("I am eatong")

    def sleep(self):
        print("I am sleeping")

    def move(self):
        print("i am moving")

class Dog(Animal):

    def __init__(self,name:str,age:int,Breed:str):
        self.breed=Breed
        super().__init__(name,age)
        print(f"Name of the animale :- {self.breed}")
        print("this is dog init")

    def bark(self):
        print("i am barking")

    def move(self):
        print("i am running on 4 legs ")

dog = Dog(name="kallu",age=56,Breed="german")
dog.bark()
dog.sleep()
dog.move()
```
