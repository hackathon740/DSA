```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def permiter(self):
        pass

class Rectangle(Shape):
    def __init__(self,length:int, breadth:int ,height:int):
        self.length = length
        self.breadth = breadth
        self.height = height

    def area(self):
        print(self.length*self.breadth)

    def permiter(self):
        print(2*(self.length+self.breadth)*self.height)

rec = Rectangle(3,4,5)
rec.area()
rec.permiter()
```
